## Paso 4 - Configurar categorias

```python
# Renombramos la columna Injury_Next_Season a Injury_Current_Season para nuestro mayor entendimiento del dato 
df_csv.rename(columns={'Injury_Next_Season': 'Injury_Current_Season'}, inplace=True)
```

```python
# Transformamos "Injury_Current_Season" y "warm up routine" a booleano (valores true /false)
df_csv['Injury_Current_Season'] = df_csv['Injury_Current_Season'].astype(bool)
df_csv['Warmup_Routine_Adherence'] = df_csv['Warmup_Routine_Adherence'].astype(bool)


print (df_csv[['Warmup_Routine_Adherence', 'Injury_Current_Season']].head(10))
```

```python
#Creamos una nueva columna con las nuevas categorías para "height_cm"
df_csv['Height_category'] = ""
df_csv.head(5)

#Sacamos los percentiles 33% y 66% de "height_cm"
height33 = int(df_csv['Height_cm'].quantile(0.33))
height66 = int(df_csv['Height_cm'].quantile(0.66))
print(f'El percentil 33% es igual a {height33} cm y el el percentil 66% es igual a {height66} cm')

#Creamos las categorías para "height_cm" y añadimos los nuevos valores en la columna "Height_category"
for i in range(len(df_csv)):
    #Si el valor de cada fila en la columna "Height_cm" es menor que 174, el valor en cada fila en la columna "Height_range" será '< 174 cm'
    if df_csv.loc[df_csv.index[i], 'Height_cm'] < height33:
        df_csv.loc[df_csv.index[i], 'Height_category'] = f'< {height33} cm'

    #Si el valor de cada fila en la columna "Height_cm" es entre 174 y 180, el valor en cada fila en la columna "Height_range" será '174 - 180 cm'
    elif df_csv.loc[df_csv.index[i], 'Height_cm'] >= height33 and df_csv.loc[df_csv.index[i], 'Height_cm'] <= height66:
        df_csv.loc[df_csv.index[i], 'Height_category'] = f'{height33} - {height66} cm'

    #Si el valor de cada fila en la columna "Height_cm" es , el valor en cada fila en la columna "Height_range" será ''
    else:
        df_csv.loc[df_csv.index[i], 'Height_category'] = f'> {height66} cm'

print(df_csv['Height_category'].head(5))
df_csv.head(10)
```

```python
#Creamos una nueva columna de categorias del sueño
df_csv['Sleep_category'] = ""
df_csv.head(5)

#Creamos las categorías para "Sleep_Hours_Per_Night" y añadimos los nuevos valores en la columna "Sleep_category"
for i in range (len(df_csv)):
    #Si la edad del futbolista es entre 18 a 20 años lo recomnedable es dormir entre 8 y 10 horas al día. 
    #Si duerme duerme menos de 8 horas, la categoría de su sueño es "Insufficient"/ si duerme entre 8 y 10 horas es "Optimal" / si duerme más de 10 horas es "Excessive
    if df_csv.loc[df_csv.index[i],'Age'] >=18 and df_csv.loc[df_csv.index[i],'Age'] <= 20:
        if df_csv.loc[df_csv.index[i],'Sleep_Hours_Per_Night'] < 8:
            df_csv.loc[df_csv.index[i],'Sleep_category'] = 'Insufficient'
        elif df_csv.loc[df_csv.index[i],'Sleep_Hours_Per_Night'] >= 8 and df_csv.loc[df_csv.index[i],'Sleep_Hours_Per_Night'] <= 10:
            df_csv.loc[df_csv.index[i],'Sleep_category'] = 'Optimal'
        else:
            df_csv.loc[df_csv.index[i],'Sleep_category'] = 'Excessive'
    #Para los futbolistas mayores de 20 años lo recomnedable es dormir entre 7 y 9 horas al día. 
    #Si duerme menos de 7 horas, la categoría de su sueño es "Insufficient" / si duerme entre 7 y 9 horas es "Optimal" / si duerme más de 9 horas es "Excessive"
    else:
        if df_csv.loc[df_csv.index[i],'Sleep_Hours_Per_Night'] < 7:
            df_csv.loc[df_csv.index[i],'Sleep_category'] = 'Insufficient'
        elif df_csv.loc[df_csv.index[i],'Sleep_Hours_Per_Night'] >= 7 and df_csv.loc[df_csv.index[i],'Sleep_Hours_Per_Night'] <= 9:
            df_csv.loc[df_csv.index[i],'Sleep_category'] = 'Optimal'
        else:
            df_csv.loc[df_csv.index[i],'Sleep_category'] = 'Excessive'

df_csv[['Age', 'Sleep_Hours_Per_Night', 'Sleep_category']]
```

```python
# Calculamos el total de lesiones en esta temporada.
injured = (df_csv['Injury_Current_Season'] == True).sum()
notInjured = (df_csv['Injury_Current_Season'] == False).sum()

fig, ax = plt.subplots(figsize=(4, 4))

# ax.pie() crea un gráfico circular.
# autopct='%1.1f%%' → muestra el porcentaje dentro de cada porción
#   %1.1f = un decimal, %% = símbolo de porcentaje literal
# startangle=90     → empieza a dibujar desde las 12 en punto (arriba)
# colors            → lista de colores para cada porción
# explode           → separa ligeramente una porción del centro (efecto visual)
#   (0.05, 0, 0, 0) separa la primera porción un 5%
# textprops         → propiedades del texto (tamaño de fuente)
colores_pie = ['#E53935','#43A047']
explode = (0.05, 0)  # Separamos ligeramente la porción más grande

wedges, texts, autotexts = ax.pie(
    [injured, notInjured],
    labels=['Injured', 'Not Injured'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colores_pie,
    explode=explode,
    textprops={'fontsize': 12}
)


# Ponemos en negrita los porcentajes para que destaquen.
for autotext in autotexts:
    autotext.set_fontweight('bold')

ax.set_title('Distribución de lesiones durante la temporada actual',
             fontsize=14, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()

# Mostramos también los valores absolutos para complementar el gráfico.
print(f"\nValores absolutos: {injured} futbolistas se lesionaron y {notInjured} futbolistas no se lesionaron")
#for ruta, llegadas in llegadas_ruta.sort_values(ascending=False).items():
  #  print(f"  {ruta}: {llegadas:,} llegadas")
```
