La columna que indica si un jugador se lesionó estaba etiquetada de forma confusa. La renombramos para que refleje con exactitud lo que mide: si el jugador sufrió una lesión durante la temporada que estamos analizando. Un nombre claro evita interpretaciones erróneas.

Las variables que nos dicen si el jugador se lesionó y si realizó el calentamiento, estaban actualmente escritas con 0s y 1s en formato Integer. Las convertimos a formato Boolean (true y false) que nos proporciona mayor claridad semántica y mejor visualización en futuros gráficos.

Dividimos a los jugadores en tres grupos según su altura (bajo, medio y alto) en relación al propio grupo. Usamos la distribución real del dataset: el tercio inferior, el central y el superior. Esto nos ayuda a simplificar el análisis, facilitar la comparación entre grupos y detectar patrones de forma más clara.

```python
#Creamos una nueva columna con las nuevas categorías para "Height_cm"
df_csv['Height_category'] = ""
df_csv.head(5)

#Sacamos los percentiles 33% y 66% de "height_cm"
height33 = int(df_csv['Height_cm'].quantile(0.33))
height66 = int(df_csv['Height_cm'].quantile(0.66))
print(f'El percentil 33% es igual a {height33} cm y el el percentil 66% es igual a {height66} cm')

#Creamos las categorías para "height_cm" y añadimos los nuevos valores en la columna "Height_category"
for i in range(len(df_csv)):
    #Si el valor de cada fila en la columna "Height_cm" es menor que 174, el valor en cada fila en la columna "Height_range" será '< 174 cm'
    if df_csv.loc[df_csv.index[i], 'Height_cm'] < height33:
        df_csv.loc[df_csv.index[i], 'Height_category'] = f'< {height33} cm'

    #Si el valor de cada fila en la columna "Height_cm" es entre 174 y 180, el valor en cada fila en la columna "Height_range" será '174 - 180 cm'
    elif df_csv.loc[df_csv.index[i], 'Height_cm'] >= height33 and df_csv.loc[df_csv.index[i], 'Height_cm'] <= height66:
        df_csv.loc[df_csv.index[i], 'Height_category'] = f'{height33} - {height66} cm'

    #Si el valor de cada fila en la columna "Height_cm" es , el valor en cada fila en la columna "Height_range" será ''
    else:
        df_csv.loc[df_csv.index[i], 'Height_category'] = f'> {height66} cm'

print(df_csv['Height_category'].head(5))
df_csv.head(10)
```

Aplicamos el mismo criterio con el peso: tres grupos iguales en función de cómo se distribuye el peso entre todos los jugadores. La consistencia en el método facilita comparar los resultados entre variables.

```python
#Creamos una nueva columna con las nuevas categorías para "Weight_kg"
df_csv['Weight_category'] = ""
df_csv.head(5)

#Sacamos los percentiles 33% y 66% de "weight_kg"
weight33 = int(df_csv['Weight_kg'].quantile(0.33))
weight66 = int(df_csv['Weight_kg'].quantile(0.66))
print(f'El percentil 33% es igual a {weight33} kg y el el percentil 66% es igual a {weight66} kg')

#Creamos las categorías para "Weight_kg" y añadimos los nuevos valores en la columna "Weight_category"
for i in range (len(df_csv)):
    #Si el valor de cada fila en la columna "Weight_kg" es menor que 69, el valor en cada fila en la columna "Weight_category" será '< 69 kg'
    if df_csv.loc[df_csv.index[i],'Weight_kg'] < weight33:
        df_csv.loc[df_csv.index[i],'Weight_category'] = f'< {weight33} kg'

    #Si el valor de cada fila en la columna "Weight_kg" es entre 69 y 78 kg, el valor en cada fila en la columna "Weight_category" será '69 - 78 kg'
    elif df_csv.loc[df_csv.index[i],'Weight_kg'] >= weight33 and df_csv.loc[df_csv.index[i],'Weight_kg'] <= weight66:
        df_csv.loc[df_csv.index[i],'Weight_category'] = f'{weight33} - {weight66} kg'

    #Si el valor de cada fila en la columna "Weight_kg" es mayor que 77 kg, el valor en cada fila en la columna "Weight_category" será '> 78 kg'
    else:
        df_csv.loc[df_csv.index[i],'Weight_category'] = f'> {weight66} kg'
df_csv.head(5)
```

El BMI (Índice de Masa Corporal) se clasifica según los rangos establecidos por la Organización Mundial de la Salud: bajo peso, peso normal, sobrepeso y obesidad. A diferencia de la altura y el peso, aquí usamos una escala médica reconocida, lo que nos permite situar a los jugadores dentro de un marco de salud general.

```python
#Creamos una nueva columna con las nuevas categorías para "BMI"
df_csv['Body_Mass_Index'] = ""
df_csv.head(5)

#Creamos las categorías para "BMI" y añadimos los nuevos valores en la columna "Body_Mass_Index"
for i in range (len(df_csv)):
    #Si el valor de cada fila en la columna "BMI" es menor que 18.5, el valor en cada fila en la columna "Body_Mass_Index" será 'Underweight'
    if df_csv.loc[df_csv.index[i],'BMI'] < 18.5:
        df_csv.loc[df_csv.index[i],'Body_Mass_Index'] = 'Underweight'

#Si el valor de cada fila en la columna "BMI" es entre 18.5 y 24.9, el valor en cada fila en la columna "Body_Mass_Index" será 'Normal weight'
    elif df_csv.loc[df_csv.index[i],'BMI'] >= 18.5 and df_csv.loc[df_csv.index[i],'BMI'] <= 24.9 :
        df_csv.loc[df_csv.index[i],'Body_Mass_Index'] = 'Normal weight'

#Si el valor de cada fila en la columna "BMI" es entre 24.9 y 29.9, el valor en cada fila en la columna "Body_Mass_Index" será 'Overweight'
    elif df_csv.loc[df_csv.index[i],'BMI'] >= 24.9 and df_csv.loc[df_csv.index[i],'BMI'] <= 29.9 :
        df_csv.loc[df_csv.index[i],'Body_Mass_Index'] = 'Overweight'

#Si el valor de cada fila en la columna "BMI" es mayor que 29.9, el valor en cada fila en la columna "Body_Mass_Index" será 'Obesity'
    else:
        df_csv.loc[df_csv.index[i],'Body_Mass_Index'] = 'Obesity'
 
df_csv[['Body_Mass_Index', 'BMI']].head(20)
```

El dataset registraba las horas de entrenamiento por semana. Las convertimos a horas diarias asumiendo que los jugadores entrenan de lunes a viernes, ya que los partidos habitualmente se juegan en fin de semana. La carga diaria es una medida más intuitiva para analizar el esfuerzo real de cada jugador.

```python
#Creamos una nueva columna con las horas de entreno diarias para "Training_Hours_Per_Week"
df_csv['Training_Hours_Per_Day'] = 0.0
df_csv.head(5)

#Asumimos que solo entrenan entre semana (Lunes a viernes) por lo tanto dividimos las horas semanales entre 5, ya que los partidos suelen ser en fin de semana
df_csv['Training_Hours_Per_Day'] = round(df_csv['Training_Hours_Per_Week']/5, 1) 

#for i in range(len(df_csv)):
    #df_csv.loc[df_csv.index[i], 'Training_Hours_Per_Day'] = round(df_csv.loc[df_csv.index[i], 'Training_Hours_Per_Week'] / 2, 1)

df_csv[['Training_Hours_Per_Day', 'Training_Hours_Per_Week']].head(10)
```

Agrupamos a los jugadores según cuántos partidos disputaron la temporada anterior: menos de 10, entre 10 y 20, entre 20 y 30, o más de 30. Un jugador con muchos partidos acumulados puede llegar a la temporada actual con una carga histórica muy diferente a uno que apenas jugó.

```python
#Creamos una nueva columna con las nuevas categorías para "Matches_Played_Past_Season"
df_csv['Matches_range_past_season'] = ""
df_csv.head(5)

#Creamos las categorías para "Matches_Played_Past_Season" y añadimos los nuevos valores en la columna "Matches_range_past_season"
for i in range (len(df_csv)):
    #Si el valor de cada fila en la columna "Matches_Played_Past_Season" es menor que 10 , el valor en cada fila en la columna "Matches_range_past_season" será 'Less than 10'
    if df_csv.loc[df_csv.index[i],'Matches_Played_Past_Season'] < 10:
        df_csv.loc[df_csv.index[i],'Matches_range_past_season'] = 'Less than 10'

#Si el valor de cada fila en la columna "Matches_Played_Past_Season" es entre 10 y 20 partidos, el valor en cada fila en la columna "Matches_range_past_season" será '10-20'
    elif df_csv.loc[df_csv.index[i],'Matches_Played_Past_Season'] >= 10 and df_csv.loc[df_csv.index[i],'Matches_Played_Past_Season'] <= 20 :
        df_csv.loc[df_csv.index[i],'Matches_range_past_season'] = '10 - 20'

#Si el valor de cada fila en la columna "Matches_Played_Past_Season" es entre 21 y 30 partidos, el valor en cada fila en la columna "Matches_range_past_season" será '21-30'
    elif df_csv.loc[df_csv.index[i],'Matches_Played_Past_Season'] >= 21 and df_csv.loc[df_csv.index[i],'Matches_Played_Past_Season'] <= 30 :
        df_csv.loc[df_csv.index[i],'Matches_range_past_season'] = '21 - 30'

#Si el valor de cada fila en la columna "Matches_Played_Past_Season" es mayor que 30 partidos, el valor en cada fila en la columna "Matches_range_past_season" será 'More than 30'
    else:
        df_csv.loc[df_csv.index[i],'Matches_range_past_season'] = 'More than 30'
df_csv.head(5)
```

```python
from plotly.subplots import make_subplots

# 🔹 1. Calculamos el porcentaje de lesión
percentageInjury = df_csv.groupby('Previous_Injury_Count')['Injury_Current_Season'].mean() * 100

# Convertimos a DataFrame
percentageInjury = percentageInjury.reset_index()
percentageInjury.columns = ['Previous_Injury_Count', 'Injury_Percentage']

# 🔹 2. Creamos la gráfica original
fig_bar = px.bar(
    percentageInjury,
    x='Previous_Injury_Count',
    y='Injury_Percentage',
    labels={
        'Previous_Injury_Count': 'Número de lesiones previas',
        'Injury_Percentage': 'Probabilidad de lesión (%)'
    }
)

fig_bar.update_traces(
    texttemplate='%{y:.1f}%',
    textposition='outside'
)

# 🔹 3. Creamos el grid 2x2
fig = make_subplots(
    rows=2,
    cols=2,
    subplot_titles=(
        "Probabilidad de lesión", "", 
        "", ""
    )
)

# 🔹 4. Añadimos la gráfica en [0,0] → (row=1, col=1)
fig.add_trace(
    fig_bar.data[0],
    row=2,
    col=2
)

# 🔹 5. Añadimos títulos de ejes (IMPORTANTE)
fig.update_xaxes(
    title_text="Número de lesiones previas",
    row=2,
    col=2
)

fig.update_yaxes(
    title_text="Probabilidad de lesión (%)",
    range=[0, 110],  # para que no se corten los %
    row=2,
    col=2
)

# 🔹 6. Layout final
fig.update_layout(
    height=700,
    width=900,
    template='plotly_white',
    showlegend=False
)

# 🔹 7. Mostrar
fig.show()
```
